<h1>Parse websites with BeautifulSoup</h1>

```python
from bs4 import BeautifulSoup
```

```python
# links of sites with relevant content
files_html = [f"file_{i}.html" for i in range(6)]
files_txt = [f"file_{i}.txt" for i in range(6)]
print(files_html)
print(files_txt)
```

```python
# write content of the links to text files
for i in range(6):
    with open(f"file_{i}.html", 'r') as f1, open(f"file_{i}.txt", 'w') as f2:
        contents = f1.read()
        soup = BeautifulSoup(contents, 'lxml')
        print('#'*79)
        print(soup.title.text)
        print(soup.body.text)
        f2.write(soup.title.text)
        f2.write(soup.body.text)
```
